## [사각형 넓이 구하기]

```python
## 사각형 넓이를 구하는 클래스
class Square:
    def __init__(self):
        self.square = int(input('** 넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n   [1.직사각형 2.평행사변형 3.사다리꼴] \n   >>>'))
        self.result = 0  # 넓이 계산 결과 출력 내용
        
        if self.square == 1:
            self.result = self.rect()
        elif self.square == 2:
            self.result = self.par()     
        elif self.square == 3:
            self.result = self.trape()   
        else:
            raise Exception('1, 2, 3 중에서 다시 입력해주세요')
    
    # 직사각형 넓이
    def rect(self):
        width, vertical = map(int, input('** 가로, 세로를 입력하세요. (예시 : 가로,세로)\n   >>>').split(','))
        area = width * vertical
        result = '[직사각형의 넓이는 : ' + str(area) +']'
        return result

    # 평행사변형 넓이
    def par(self):
        width, vertical = map(int, input('** 밑변, 높이를 입력하세요. (예시 : 밑변,높이)\n   >>>').split(','))
        area = width * vertical
        result = '[평행사변형의 넓이는 : ' + str(area) +']'
        return result

    # 사다리꼴 넓이
    def trape(self):
        top, bottom, vertical = map(int, input('** 윗변,아랫변,높이를 입력하세요. (예시 : 윗변,아랫변,높이)\n   >>>').split(','))
        area = (top + bottom) * vertical / 2
        result = '[평행사변형의 넓이는 : ' + str(area) +']'
        return result

def area():
    try:    
        # 사각형 넓이 클래스 객체 생성
        sq = Square()
       
        # 출력
        print(sq.result)       
             
    except ValueError:
        print("잘못 입력하셨습니다. 가이드대로 입력해 주세요.")
    except Exception as e:
        print(e)    
```

## [키오스크 음료 주문]

```python
## 키오스크
##    - 메뉴출력/주문/추가주문/지불/영수증출력

class Kiosk:
    menu = [[], []]        # 메뉴 목록
    order_menu = [[], []]  # 주문 메뉴
    price_sum = 0          # 총 주문금액
    pay_method = 0         # 지불수단(1:현금 2:카드)
    # 출력 포맷용
    NAME_COL  = 15
    PRICE_COL = 15

    ################
    # 초기화
    def __init__(self):
        self.menu = [['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte'] 
                    ,[2000, 3000, 3000, 2500, 2500, 3000]]

    ################
    # 라인 출력 : 메뉴 상하 외곽 라인 공통 출력
    def line(fn):
        def line_prt(*args):
            print("-" * 40) 
            print(f"      Menu                    Price  ")
            print("-" * 40) 
            fn(*args)               
            print("-" * 40)    
        return line_prt  

    ################
    # 메뉴 출력 
    @line
    def menu_print(self):
        for i in range(len(self.menu[0])):
            print(f" {i + 1} {self.menu[0][i]:<{self.NAME_COL}}  {self.menu[1][i]:>{self.PRICE_COL},} 원")

    ################
    # 주문 / 추가주문 
    def menu_select(self):
        print()
        n = -1
        while n != 0:
            # 최초주문
            if len(self.order_menu[0]) == 0:
                while n < 1 or len(self.menu[0]) < n:
                    # 첫주문 메뉴 입력
                    n = int(input("음료 번호를 입력하세요 : ")) 
                    
                    # 메뉴판에 있는 음료 번호일때 주문추가
                    if 1 <= n & n <= len(self.menu[0]):
                        self.menu_add(n)   
                        print(f"[주문 음료]         {self.order_menu[0][len(self.order_menu[0])-1]} : {self.order_menu[1][len(self.order_menu[1])-1]:,} 원")
                        
                    # 메뉴판에 없는 번호일 때
                    else:  
                        print("!! 없는 메뉴입니다. 다시 주문해 주세요.")
            
            # 추가주문 또는 지불
            else:
                while n != 0:
                    print()
                    n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))  # 추가주문 또는 지불
                    
                    # 추가메뉴주문
                    if n > 0 and n < len(self.menu[0]) + 1:
                        self.menu_add(n)
                        print(f"[추가 주문 음료]       {self.order_menu[0][len(self.order_menu[0])-1]}  {self.order_menu[1][len(self.order_menu[1])-1]:,}원")
                        print(f"[합계]  :  {self.price_sum:,} 원")
                    
                    # 지불선택
                    else:
                        if n == 0:
                            print("주문이 완료되었습니다.")
                            break
                        else:
                            print("!! 없는 메뉴입니다. 다시 주문해 주세요.")
            if n == 0: 
                break;
                
    ################  
    # 주문 메뉴 추가
    def menu_add(self, n):
        # 음료 온도 물어보기
        t = 0  
        # 1이나 2를 입력할 때까지 물어보기
        while t != 1 and t != 2:  
            t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
            # "HOT" 또는 "ICE"          
            if t == 1:
                temp = 'HOT'
            elif t == 2:
                temp = 'ICE'    
            else:    
                print("!! 1과 2 중 하나를 입력하세요.\n")

        # 주문음료 정보저장
        self.order_menu[0].append(temp + ' ' + self.menu[0][n-1]) # 온도+메뉴명
        self.order_menu[1].append(self.menu[1][n-1])            # 가격
        self.price_sum += self.menu[1][n-1]                   # 총 합계금액

    ################
    # 지불 
    def pay(self):
        # 합계금액
        print(f"\n총 합계 금액은 {self.price_sum:,}원 입니다.")

        # 지불수단 선택
        while self.pay_method != 1 and self.pay_method != 2:
            self.pay_method = int(input("현금 결제는 1번, 카드 결제는 2번을 입력해 주세요. : "))
            if self.pay_method == 1:
                print('직원을 호출하겠습니다.\n')
            elif self.pay_method == 2:
                print('IC 방향에 맞게 카드를 꽂아주세요.\n')
            else:
                print('!! 다시 결제를 시도해 주세요.')	
          
        # 주문서 출력
        print("\033[1mReceipt\033[0m")
        self.table()

    ################
    # 영수증 출력 
    @line
    def table(self):       
        # 현재시간 출력
        from datetime import datetime as t
        now = t.now()
        
        # 주문 상품명 : 해당 금액
        for i in range(len(self.order_menu[0])):
            print(f"  {self.order_menu[0][i]:<{self.NAME_COL}}  {self.order_menu[1][i]:>{self.PRICE_COL},} 원")
        
        print("-" * 40) 
        print(f"{'    Total':<{self.NAME_COL+2}}  {self.price_sum:>{self.PRICE_COL},} 원")
        print(f"{now.strftime('%Y-%m-%d %H:%M:%S'):>37}")


################        
# 주문시작 	
try:
    ord = Kiosk()
    ord.menu_print()
    ord.menu_select()
    ord.pay()
except Exception:
    print(f"!! 잘못 입력하셨습니다. 처음부터 다시 주문해 주세요.")
```
